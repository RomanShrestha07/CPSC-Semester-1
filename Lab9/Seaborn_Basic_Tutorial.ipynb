## Histogram

Histograms represent the data distribution by forming bins along with the range of the data and then drawing bars to show the number of observations that fall in each bin.In Seaborn we use histplot() function to plot histograms.Here is an example:

```python
sns.histplot(tips['total_bill'])
```

A bin in a histogram is the block that you use to combine values before getting the frequency. For instance, if you were creating a histogram of age, the bins might be 0-5, 6-10, 11-15, and so on up to 85 and up (or whatever).

```python
# To remove the curve and increase the bin size to get more detailed histogram
sns.histplot(tips['total_bill'], bins = 30)
# Add labels and title from the matplotlib library
plt.ylabel("Frequency of Occurence")
plt.title("Histogram Plot of tips['total_bill']")
```

## jointplot

Joint Plot draws a plot of two variables with bivariate and univariate graphs. We can also draw a Regression Line in Scatter Plot. Here are few of the examples of a joint plot

```python
sns.jointplot(x = 'total_bill', y ='tip', data = tips, kind = 'scatter')
```

```python
# Another way of plotting the same plot
# fit a regression line
sns.jointplot(x = tips['total_bill'], y = tips['tip'], kind = 'reg')
```

```python
# Change color
# add a linear regression fit
sns.jointplot(x = 'total_bill',y = 'tip', data = tips, kind = 'reg', color = 'red')
```

## ScatterPlot

Another way to get only the middle scatter plot of the jointplot is using lmplot which plots a simple linear relationship between two variables:

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, fit_reg = False, palette = "Set1")
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips,
           #fit_reg = False, # regression line
           hue = 'sex')   # Color by gender
```

# Box Plot

The box plot, also called the box and whisker diagram is used for depicting groups of numerical data through the quartiles. It is known as the box and whisker diagram because it is composed of a box and whiskers. Boxplot is also used for detecting the outlier in a data set.

A box plot is composed of a summary of 5 different data points: the minimum, first quartile, median, third quartile, and maximum.
- Minimum
- First Quartile or 25%
- Median (Second Quartile) or 50%
- Third Quartile or 75%
- Maximum

```python
sns.boxplot(x="day", y="total_bill", data=tips)
```

- The bottom black horizontal line of the box plot is the minimum value
- The first black horizontal line of the rectangle shape of the box plot is the first quartile
 or 25%
- The second black horizontal line of the rectangle shape of the box plot is Second quartile or 50% or median.
- The third black horizontal line of rectangle shape of the same box plot is third quartile or 75%
- The top black horizontal line of the rectangle shape of the box plot is the maximum value.
- The small diamond shape of the box plot is outlier data.

## Pairplot

pairplot will plot pairwise relationships across an entire dataframe (for the numerical columns) and supports a color hue argument (for categorical columns).

Pariplot gonna show pariwise relationshop across your entire dataframe

Basically pairplots join plot for every single combination of numeric columns in the dataframe

```python
sns.pairplot(tips)
```

Draw scatterplots for joint relationships and histograms for univariate distributions:
